```python
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from tools import model_predict, text_generation
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/tolik/information_technology/third_year/practice_project/CoreAnalysis-ML/data_for_study/data.csv')
model_uf = torch.load('DeepLabV3Plus_resnet18_ultraviolet.pth')
model_ds = torch.load('R2AttUnet_daylight.pth')
```

```python
classes = ['Переслаивание пород', 'Алевролит глинистый', 'Песчаник',
               'Аргиллит', 'Проба', 'Разлом']
for id_mask in [1007202, 1017348, 1006162, 1000344]:
    orig_mask = np.load(f'C:/Users/tolik/information_technology/third_year/practice_project/CoreAnalysis-ML/data_for_study/labels/daylight/label_{id_mask}.npz')['x']
    orig_img = Image.open(f'C:/Users/tolik/information_technology/third_year/practice_project/CoreAnalysis-ML/data_for_study/photos/{id_mask}.jpeg')
    predict = model_predict(model_ds, orig_img)
    
    res = text_generation(predict, 'daylight')
    for i in res:
        print(i)
    
    
    class_indexes = np.unique(orig_mask)
    current_classes = []
    for i in class_indexes:
        current_classes.append(classes[i])
    print('В оригинальной маске классы: ', current_classes)
    
    
    ax1 = plt.subplot(131)
    plt.imshow(orig_img)
    ax1.set_title(f'orig photo {id_mask}')
    
    ax2 = plt.subplot(132)
    plt.imshow(predict)
    ax2.set_title('predict mask')
    
    ax3 = plt.subplot(133)
    plt.imshow(orig_mask)
    ax3.set_title(f'orig mask')
    
    
    display(data[data.photo_id==id_mask])
    
    plt.pause(0.001)
```
