```python
import skimage.segmentation as seg
import matplotlib.pyplot as plt
import cv2

IMAGE_PATH = 'C:/Users/tolik/information_technology/third_year/practice_project/CoreAnalysis-ML/data_for_study/photos/'
ULTRA_MASK_PATH = 'C:/Users/tolik/information_technology/third_year/practice_project/CoreAnalysis-ML/data_for_study/labels/ultraviolet/label_'
```

```python
def image_show_id(img_id, ncols=1, nrows=1):
    img = cv2.imread(IMAGE_PATH + img_id + '.jpeg')
    fig, ax = plt.subplots(nrows, ncols, figsize=(25,10))
    ax.imshow(img, cmap='gray')
    ax.axis('off')
    return fig, ax
def image_show_file(img_file, ncols=1, nrows=1):
    fig, ax = plt.subplots(nrows, ncols, figsize=(25,10))
    ax.imshow(img_file, cmap = 'gray')
    ax.axis('off')
    return fig, ax
```

Гистограмма - это график, на котором ось Х показывает все возможные значения пикселей, а ось Y показывает частоту появления этих значений на изображении

```python
image = cv2.imread(IMAGE_PATH + '1011988.jpeg')
image_show_id('1011988')
```

Контролируемый порог

```python
image_seg = (image < 160).astype(float)
image_show_file(image_seg)
```

Неконтролируемый порог

```python
import skimage.filters as filt
image_threshold_li = filt.threshold_li(image)
image_threshold_otsu = filt.threshold_otsu(image)
#image_threshold_local = filt.threshold_local(image_gray, block_size=31, method='mean', offset=10)
image_show_file((image < image_threshold_li).astype(float))
image_show_file((image < image_threshold_otsu).astype(float))
#image_show_file((image_gray < image_threshold_local).astype(float))
```
